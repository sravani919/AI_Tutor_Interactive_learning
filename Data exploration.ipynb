```python
import pandas as pd
import random

# Load the dataset
file_path = "final_review_questions_cleaned.csv"  # Ensure the path is correct
df = pd.read_csv(file_path)

# Step 1: Remove invalid questions that contain AI-generation text
invalid_patterns = [
    "Expand the following short answer", 
    "Generated answer needed",
    "Short Answer:",
    "Expanded Answer:"
]

# Remove rows where the question contains AI-generated prompt text
df = df[~df["question"].astype(str).str.contains("|".join(invalid_patterns), na=False)]

# Function to create multiple-choice quizzes
def generate_quiz(df, num_questions=5):
    quiz_questions = []

    for _ in range(num_questions):
        # Randomly select a question
        question_row = df.sample(1).iloc[0]  
        question_text = question_row["question"]
        correct_answer = question_row["answer"]

        # Limit the correct answer length to avoid very long responses
        correct_answer = correct_answer[:150] + "..." if len(correct_answer) > 150 else correct_answer

        # Get 3 random incorrect answers
        incorrect_answers = df[df["answer"] != correct_answer]["answer"].sample(3).tolist()

        # Ensure incorrect answers are not too long
        incorrect_answers = [ans[:150] + "..." if len(ans) > 150 else ans for ans in incorrect_answers]

        # If we don't have enough incorrect answers, fill with placeholder text
        while len(incorrect_answers) < 3:
            incorrect_answers.append("Random Choice")

        # Shuffle the options
        options = [correct_answer] + incorrect_answers
        random.shuffle(options)

        # Store the formatted quiz question
        quiz_questions.append({
            "Question": question_text,
            "Options": options,
            "Correct Answer": correct_answer
        })

    return quiz_questions

# Generate a quiz with 5 questions
quiz = generate_quiz(df, num_questions=5)

# Display the quiz
for i, q in enumerate(quiz, 1):
    print(f"\n📌 Question {i}: {q['Question']}")
    for idx, option in enumerate(q["Options"], 1):
        print(f"  {idx}. {option}")
    print(f"✅ Correct Answer: {q['Correct Answer']}\n")
```
